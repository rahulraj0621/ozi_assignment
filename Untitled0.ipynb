```python
users = pd.read_csv('users.csv')
orders = pd.read_csv('orders.csv')
events = pd.read_csv('order_events.csv')
feedback = pd.read_csv('feedback.csv')
```

```python
events['event_timestamp'] = pd.to_datetime(events['event_timestamp'])
```

```python
pickups = events[events['event_status'] == 'PICKED_UP'][['order_id', 'event_timestamp']]
deliveries = events[events['event_status'] == 'DELIVERED'][['order_id', 'event_timestamp']]
```

```python
duration_table = pd.merge(pickups, deliveries, on='order_id', suffixes=('_pick', '_drop'))
```

```python
duration_table = pd.merge(pickups, deliveries, on='order_id', suffixes=('_pick', '_drop'))
```

```python
duration_table.head()
```

```python
master = pd.merge(orders, duration_table, on='order_id', how='inner')
```

```python
master = pd.merge(master, feedback, on='order_id', how='left')
```

```python
master = pd.merge(master, users, on='user_id', how='left')
```

```python
master['order_date'] = pd.to_datetime(master['order_date'])
```

```python
master.head()
```

```python
threshold_value = master['order_amount'].mean()
print(f"High Value Threshold (Mean): {threshold_value:.2f}")
```

```python
jan_orders = master[master['order_date'].dt.month == 1]
```

```python
jan_orders.head()
```

```python
master['actual_time_mins'] = (master['event_timestamp_drop'] - master['event_timestamp_pick']).dt.total_seconds() / 60
jan_orders = master[master['order_date'].dt.month == 1]
```

```python
happy_high_value_jan = jan_orders[
    (jan_orders['order_amount'] > threshold_value) &
    (jan_orders['rating'] >= 4) &
    (jan_orders['actual_time_mins'] <= jan_orders['promised_time_mins'])]
```

```python
vip_jan_users = happy_high_value_jan['user_id'].unique()
print(f"Happy VIP Users in Jan: {len(vip_jan_users)}")
```

```python
feb_orders = master[master['order_date'].dt.month == 2]
```

```python
returning_vips = feb_orders[feb_orders['user_id'].isin(vip_jan_users)]['user_id'].unique()
```

```python
print(f"Returning_vips in Feb: {len(returning_vips)}")
```

```python
retention_rate = (len(returning_vips) / len(vip_jan_users)) * 100
```

```python
print(f"Retention Rate of Happy High-Value Users: {retention_rate:.2f}%")
```

```python
avg_fleet_time = master['actual_time_mins'].mean()
```

```python
print(f"Average Fleet Delivery Time: {avg_fleet_time:.2f} minutes")
```
